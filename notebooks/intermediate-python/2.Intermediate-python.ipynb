## Matplotlib

- matplotlib is a python library for display grafics or plot

```python
#visualizing evolution of world's population
import matplotlib.pyplot as plt 
year = [1950, 1970 , 1990, 2010]
population = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, population)  # plt.plot(horizontal_axis, vertical_axis)  
plt.show()
```

## Scatter Plot
- Scatter plot is a data visualization tool that uses dots on a Cartesian coordinate plane (x and y axes) to represent values for two different numeric variables

```python
import matplotlib.pyplot as plt 
year = [1950, 1970 , 1990, 2010]
population = [2.519, 3.692, 5.263, 6.972]
plt.scatter(year, population)  # plt.plot(horizontal_axis, vertical_axis)  
plt.show()
```

## Histsogram

- A type of visualization useful for exploring dataset.
- It helps to get idea about distribution of variables.

- x argument is the list of values that we want to build the histogram for and bins argument is used to tell for how many bins the data should be divided and by default it is 10.

```python
values = [1, 1.1, 2.6, 3.5, 1.3, 0.5, 0.9, 1.9, 2.3, 3.2, 3.4, 2.1]
plt.hist(values, bins = 3)
plt.show()
```

### Exercise

```python
import numpy as np

# sample data distribution of exam scores
exam_scores = np.random.normal(75, 10, 1000)  # 1000 students scores with mean=75, std=10  --> follows normal distribution

# Create histogram of exam_scores data
plt.hist(exam_scores, bins=20, color='skyblue', edgecolor='black')

# Display histogram
plt.show()
```

```python
# Build histogram with 5 bins
plt.hist(exam_scores, bins=5, color='skyblue', edgecolor='black')

# Show and clean up plot
plt.show()
plt.clf()    # cleans up the plot and start afresh

# Build histogram with 20 bins
plt.hist(exam_scores, bins=20, color='pink', edgecolor='black')

# Show and clean up again
plt.show()
plt.clf()
```

- Here we can see too many bins will overcomplicate the reality and hide the bigger picture.
- plt.clf() helps to clear up the plot and start afresh.

## Customization

```python
# Years from 1950 to 2100
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2025, 2030, 
                  2040, 2050, 2060, 2070, 2080, 2100]

# Estimated and projected population (in billions)
population = [2.5, 3.0, 3.7, 4.4, 5.3, 6.1, 6.9, 7.8, 8.1, 8.5, 
                       9.1, 9.7, 10.0, 10.2, 10.4, 10.2]

years = [1800, 1850, 1900] + years
population = [1.0, 1.262, 1.650] + population

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(years, population, color='b', label='Population (Billions)')

# Adding titles and labels
plt.title('Global Population Growth (1950-2100)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks([0, 2, 4, 6, 8, 10],               # this changes the y coordinate accordingly
           ['0', '2B', '4B', '6B', '8B', '10B'])

plt.legend()
plt.show()
```

```python
# Sample data: advertising budget (in 1000)
ad_budget = np.random.normal(700, 10, 300)  # Advertising budget
sales_revenue = np.random.normal(300, 10, 300) # Corresponding sales revenue

# Basic scatter plot, log scale
plt.scatter(ad_budget, sales_revenue, color='blue', alpha=0.7)
plt.xscale('log') 

# Strings for labels and title
xlab = 'Advertising Budget [in $1000]'
ylab = 'Sales Revenue [in $1000]'
title = 'Impact of Advertising Budget on Sales Revenue'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

# Display the plot
plt.show()
```

- plt.yticks([0, 1, 2], ["Low", "Medium", "High"]) and here yticks will help to replace 0, 1, and 2 with Low, Medium and High

```python
# Sample data: advertising budget (in 1000)
ad_budget = np.random.normal(700, 10, 300)  # Advertising budget
sales_revenue = np.random.normal(300, 10, 300) # Corresponding sales revenue

# Basic scatter plot, log scale
plt.scatter(ad_budget, sales_revenue, color='blue', alpha=0.7)
plt.xscale('log') 

# Strings for labels and title
xlab = 'Advertising Budget [in $1000]'
ylab = 'Sales Revenue [in $1000]'
title = 'Impact of Advertising Budget on Sales Revenue'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

tick_val = [600, 650, 700, 750]
tick_lab = ['0.6k', '0.65k', '0.7k', '0.75']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)

plt.show()
```

```python
# Sample data: marketing budget (in 1000)
marketing_budget = [1, 5, 10, 50, 100, 500, 1000, 5000, 10000]  
sales_revenue = [5, 20, 35, 80, 150, 400, 700, 3000, 6000]  

# Store number of campaigns as a numpy array: np_campaigns
np_campaigns = np.array([10, 30, 50, 80, 100, 200, 300, 500, 800])

# Double np_campaigns to emphasize marker sizes
np_campaigns *= 2

# Create scatter plot with varying marker sizes
plt.scatter(marketing_budget, sales_revenue, s=np_campaigns, alpha=0.7)

# Customize axes
plt.xscale('log') 
plt.xlabel('Marketing Budget [in $1000]')
plt.ylabel('Sales Revenue [in $1000]')
plt.title('Marketing Budget vs Sales Revenue')

# Custom x-axis ticks
plt.xticks([10, 100, 1000, 10000], ['10k', '100k', '1M', '10M'])

# Display the plot
plt.show()
```

```python
# Sample colors for each data point
colors = ['red', 'blue', 'green', 'purple', 'orange', 'pink', 'brown', 'cyan', 'yellow']

# Create scatter plot with color and transparency
plt.scatter(x=marketing_budget, y=sales_revenue, s=np_campaigns, c=colors, alpha=0.8)

# Customize axes
plt.xscale('log') 
plt.xlabel('Marketing Budget [in $1000]')
plt.ylabel('Sales Revenue [in $1000]')
plt.title('Marketing Budget vs Sales Revenue')

# Custom x-axis ticks
plt.xticks([10, 100, 1000, 10000], ['10k', '100k', '1M', '10M'])

# Show the plot
plt.show()
```
